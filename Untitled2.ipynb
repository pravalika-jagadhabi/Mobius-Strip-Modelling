```python
import numpy as np
import matplotlib.pyplot as plt

class MobiusStrip:
    def __init__(self, R, w, n):
        """
        Initialize the MobiusStrip object.

        Parameters:
        R (float): Radius of the strip (distance from center)
        w (float): Width of the strip
        n (int): Number of points used for creating the mesh grid
        """
        self.R = R
        self.w = w
        self.n = n

        # Create arrays
        self.u_values = np.linspace(0, 2 * np.pi, n)
        self.v_values = np.linspace(-w / 2, w / 2, n)

        # Calculate the mesh grid
        self.X, self.Y, self.Z = self.calculate_mesh()

    def calculate_mesh(self):
        """
        Calculate X, Y, Z coordinates for each point on the Mobius strip
        using the parametric equations.

        Returns:
        Three 2D numpy arrays representing X, Y, Z coordinates.
        """
        X = np.zeros((self.n, self.n))
        Y = np.zeros((self.n, self.n))
        Z = np.zeros((self.n, self.n))

        for i in range(self.n):
            u = self.u_values[i]
            cos_u = np.cos(u)
            sin_u = np.sin(u)
            cos_u_half = np.cos(u / 2)
            sin_u_half = np.sin(u / 2)

            for j in range(self.n):
                v = self.v_values[j]
                # Parametric equations of Mobius strip
                X[i, j] = (self.R + v * cos_u_half) * cos_u
                Y[i, j] = (self.R + v * cos_u_half) * sin_u
                Z[i, j] = v * sin_u_half

        return X, Y, Z

    def partial_derivative_u(self, u, v):
        """
        Calculate partial derivatives of position vector w.r.t u based on parametric equations.

        Returns:
        A numpy array [dx/du, dy/du, dz/du]
        """
        du = 1e-5  # small step to approximate derivative numerically
        f1 = self.point(u + du, v)
        f0 = self.point(u, v)
        der = (f1 - f0) / du
        return der

    def partial_derivative_v(self, u, v):
        """
        Calculate partial derivatives of position vector w.r.t v based on parametric equations.

        Returns:
        A numpy array [dx/dv, dy/dv, dz/dv]
        """
        dv = 1e-5  # small step to approximate derivative numerically
        f1 = self.point(u, v + dv)
        f0 = self.point(u, v)
        der = (f1 - f0) / dv
        return der

    def point(self, u, v):
        """
        Compute a point (x,y,z) on the Mobius strip for given u and v.

        Returns:
        numpy array [x, y, z]
        """
        x = (self.R + v * np.cos(u / 2)) * np.cos(u)
        y = (self.R + v * np.cos(u / 2)) * np.sin(u)
        z = v * np.sin(u / 2)
        return np.array([x, y, z])

    def surface_area(self):
        """
        Approximate the surface area of the Mobius strip using numerical integration.

        Returns:
        Approximate surface area (float).
        """
        total_area = 0.0
        # We'll integrate over the mesh indices (i,j) as representatives of u and v

        for i in range(self.n - 1):
            for j in range(self.n - 1):
                u = self.u_values[i]
                v = self.v_values[j]

                # Calculate tangent vectors at (u,v)
                du_vec = self.partial_derivative_u(u, v)
                dv_vec = self.partial_derivative_v(u, v)

                # Area element is length of cross product of partial derivatives
                cross = np.cross(du_vec, dv_vec)
                dA = np.linalg.norm(cross) * (self.u_values[1] - self.u_values[0]) * (self.v_values[1] - self.v_values[0])

                total_area += dA

        return total_area

    def edge_length(self):
        """
        Numerical approximation of Mbius strip edge length (boundary curve where v = -w/2).

        Returns:
        Approximate edge length (float).
        """
        length = 0.0
        v_edge = -self.w / 2

        for i in range(self.n - 1):
            p1 = self.point(self.u_values[i], v_edge)
            p2 = self.point(self.u_values[i + 1], v_edge)
            segment = np.linalg.norm(p2 - p1)
            length += segment

        # The strip has two edges: at v = -w/2 and v = +w/2, sum both lengths
        length_other_edge = 0.0
        v_edge_other = self.w / 2
        for i in range(self.n - 1):
            p1 = self.point(self.u_values[i], v_edge_other)
            p2 = self.point(self.u_values[i + 1], v_edge_other)
            segment = np.linalg.norm(p2 - p1)
            length_other_edge += segment

        return length + length_other_edge

    def plot(self):
        """
        Plot the Mbius strip using matplotlib 3D plotting.
        """
        fig = plt.figure(figsize=(8, 6))
        ax = fig.add_subplot(111, projection='3d')
        ax.plot_surface(self.X, self.Y, self.Z, rstride=1, cstride=1, color='lightblue', edgecolor='gray', alpha=0.8)
        ax.set_title('Mbius Strip')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        ax.set_zlabel('Z')
        plt.tight_layout()
        plt.show()


if __name__ == "__main__":

    R = 1.0
    w = 0.3
    n = 50

    # Create Mobius strip object
    mobius = MobiusStrip(R, w, n)

    # Calculate surface area and edge length
    area = mobius.surface_area()
    length = mobius.edge_length()

    print(f"Approximate Surface Area: {area:.4f}")
    print(f"Approximate Edge Length: {length:.4f}")

    # Visualize the Mobius strip
    mobius.plot()
```
